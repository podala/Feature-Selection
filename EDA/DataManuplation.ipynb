#fillna- filling the missing value

mean,median,mode- numerical data
mode- Categorical data

```python
data["Item_Weight"].fillna(data["Item_Weight"].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data["Outlet_Size"].unique
```

```python
data["Outlet_Size"].mode()
```

```python
data["Outlet_Size"].fillna(data["Outlet_Size"].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
data["Outlet_Location_Type"].fillna(data["Outlet_Location_Type"].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

Slicing/selecting

--loc get rows(or columns ) with partciular label from the index
---iloc--get rows(or columns ) with partciular position in the index

```python
data["Item_Fat_Content"].head(5) # selecting particular single column
```

```python
data[["Item_Weight","Profit"]] # selecting multiple columns
```

```python
data.loc[0:10,:]             # accept 2 integer value 1 for row and 1 for column
                              #: means all the columns
```

```python
data.loc[0:10,["Item_Weight","Profit"]]
```

```python
data.loc[[1,3,5,7],["Item_Weight","Profit"]] # access throough particular row label
```

```python
data.iloc[:,0:3]   # all rows of first 3 column
```

```python
# i need row no.1,3,5,7  ,outlet_size,item type,profit, using iloc
data.iloc[[1,3,5,7],[8,4,12]]
```

```python
# Filtering
data[data["Item_Weight"]>10] # filter w.r.t to some condition
```

```python
data.loc[data["Item_Weight"]>10,["Item_Type","Profit"]]
```

```python
# find the item type in Lowfat category

data.loc[data["Item_Fat_Content"]=="Low Fat",["Item_Type","Item_MRP"]]
```
